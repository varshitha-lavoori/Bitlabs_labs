```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data={'Difficulty':['easy','medium','hard','medium','easy'],'Test Score':[50,60,70,80,90]}
#creating dataframe
df=pd.DataFrame(data)
#initialize the labelencoder
encoder=LabelEncoder()
#apply label encodei to difficulty column
df['diff']=encoder.fit_transform(df['Difficulty'])
print(df)
corr=df['diff'].corr(df['Test Score'])
print("correlation:",corr)
```

```python
import pandas as pd
data={'color':['red','blue','green','blue','red'],'test':[1,2,3,4,5]}
df=pd.DataFrame(data)
#applying one-hot coding
encode=pd.get_dummies(df,columns=['color'])
print(encode)
cor=encode['color_red'].corr(encode['test'])
print(cor)
```
