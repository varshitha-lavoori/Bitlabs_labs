```python
import numpy as np
data["MILES_CAT"]=np.where(data['MILES*']>100,"Long trip","short trip")#to create new wcolumn
data.head()
```

```python
# creating a new column
data['new_col']=10
data
```

```python
data['MILES_DOG'] = np.select(
    [
        data['MILES*'] <= 100,
        (data['MILES*'] > 100) & (data['MILES*'] < 200),
        data['MILES*'] >= 200
    ],
    ['short trip', 'medium trip', 'long trip'],
    default='unknown'
)
data.head()
```

```python
# count of each trip
cou=data['MILES_DOG'].value_counts()
cou
```

```python
# average miles of each purpose
grouped=data.groupby('PURPOSE*')['MILES*'].agg(Mean='mean')
grouped
```
