```python
#Load the dataset
import pandas as pd
data=pd.read_csv("Uber.csv")
data
```

```python
#Display basic info about dataset
data.info()
```

```python
#Check for missing values
data.isnull().sum()
```

```python
#Drop rows with missing values
data.dropna()
```

```python
#fill missing values (propose column with unknown value)
data['PURPOSE*'].fillna("Unknown")
```

```python
#Check and remove duplicates
data.duplicated()#used to check duplicate values prsnt or not

```

```python
data.drop_duplicates()#to remove duplicates
```

```python
#Convert START_DATE and END_DATE to datetime
data['START_DATE*'] = pd.to_datetime(data['START_DATE*'], errors='coerce')
data['END_DATE*'] = pd.to_datetime(data['END_DATE*'], errors='coerce')
data.dtypes
```

```python
#Total number of rides per category:
rides=data['CATEGORY*'].value_counts()
print(rides)
```

```python
#Total miles traveled for each purpose:
total_miles=data.groupby('PURPOSE*')['MILES*'].sum()
print(total_miles)
```

```python
#Average distance for business vs. personal rides:
average=data.groupby('CATEGORY*')['MILES*'].mean()
print(average)
```

```python
#Add a column for cost estimation (assuming $2 per mile):
data['Cost_estimate*'] = data['MILES*'] * 2
print(data)
```

```python
#Filter rides longer than 50 miles:
long=data[data['MILES*']>50]
print(long)
```

```python
#Filter by specific purpose (e.g., meetings):
meeting=data[data['PURPOSE*']=='Meeting']
meeting
```

```python
#What is the total number of business trips versus personal trips?
total_num=data['CATEGORY*'].value_counts()
total_num
```

```python
#What percentage of trips are business versus personal?
total=data['CATEGORY*'].value_counts()
perc=(total/total.sum())*100
perc
```
